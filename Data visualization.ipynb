```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
A = [21, 22, 23, 24, 25]
B = [22, 24, 21, 25, 23]
C = [1, 2, 3, 4, 5]
```

```python
plt.scatter(C, A)
```

```python
plt.scatter(C, A, marker = 'D', color = 'purple', label = 'Person A')
plt.scatter(C, B, marker = 'h', color = 'red', label = 'Person B')

# the following stays the same no matter what type is shown like bar or box, etc.
# marker is shape, grid is background lines, legend is info about graphs 1 and 2, which displays labels
plt.xlabel('Months')
plt.ylabel('Sales Performance')
plt.title('no. of products sold')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.boxplot(C, A, label = 'Person A')
plt.boxplot(C, B, label = 'Person B')

# the following stays the same no matter what type is shown like bar or box, etc.
plt.xlabel('Months')
plt.ylabel('Sales Performance')
plt.title('no. of products sold')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.bar(C, A, color = 'purple', label = 'Person A')
plt.bar(C, B, color = 'yellow', label = 'Person B')

# the following stays the same no matter what type is shown like bar or box, etc.
plt.xlabel('Months')
plt.ylabel('Sales Performance')
plt.title('no. of products sold')
plt.grid()
plt.legend()
plt.show()
```

```python
plt.subplot(121)   #1 row 2 columns 1st graph
plt.scatter(C, A)
plt.xlabel('Months')
plt.ylabel('Sales Performance')
plt.title('A -no. of products sold')
plt.grid()
plt.legend(['Person A'])

plt.subplot(122)   #1 row 2 columns 2nd graph
plt.scatter(C, B)
plt.xlabel('Months')
plt.ylabel('Sales Performance')
plt.title('B -no. of products sold')
plt.grid()
plt.legend(['Person B'])

plt.show()
```

```python
# scatter plot, line plot, area i.e. stack plot, bubble chart, heat maps, etc.
```

```python
plt.subplot(221)   #1 row 2 columns 1st graph
plt.scatter(C, A)
plt.xlabel('Months')
plt.ylabel('Sales Performance')
plt.title('A -no. of products sold')
plt.grid()
plt.legend(['Person A'], loc = 'upper left')

plt.subplot(222)   #1 row 2 columns 1st graph
plt.plot(C, A, linestyle = 'dotted', color = 'red')
plt.xlabel('Months')
plt.ylabel('Sales Performance')
plt.title('A -no. of products sold')
plt.grid()
plt.legend(['Person A'], loc = 'upper right')


plt.subplot(223)   #1 row 2 columns 2nd graph
plt.scatter(C, B)
plt.xlabel('Months')
plt.ylabel('Sales Performance')
plt.title('B -no. of products sold')
plt.grid()
plt.legend(['Person B'],loc = 'lower left')

plt.subplot(224)   #1 row 2 columns 2nd graph
plt.plot(C, B, color = 'black')
plt.xlabel('Months')
plt.ylabel('Sales Performance')
plt.title('B -no. of products sold')
plt.grid()
plt.legend(['Person B'],loc = 'lower right')

plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
plt.show()
```
