```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [_ for _ in range(0, 10)]
y = [_*_ for _ in range(0, 10)]

fig = plt.figure()
ax = fig.gca()

ax.set_xticks(np.arange(0, 10, 1))
ax.set_yticks(np.arange(0, 100, 10))

plt.scatter(x, y)
plt.grid()
plt.show()
```

```python
plt.plot([x for x in range (0, 10)], marker = 'p')
plt.show()
```

```python
plt.plot([x for x in range(0, 10)], marker = 'D')
plt.show()
```

```python
plt.plot([x * x for x in range(0, 10)],'red', marker = 'p')
plt.plot([x * x for x in range(9, -1, -1)],'blue', marker='p')
plt.show()
```

```python
plt.title("Cube of nums from 0 to 9", fontname = 'Comic Sans MS', fontsize = 18)
plt.xlabel("nums from 0 to 9", fontname = 'Arial', fontsize = 12)
plt.ylabel("cube of nums on x axis", fontname = 'sans-serif', fontsize = 12)
plt.plot([x for x in range(0, 10)], 'cyan', marker = 'p')
plt.show()
```

## Assignment 3

```python
sns.histplot(x = 'wt', data = df1, bins = 30, kde = True, color = 'skyblue')
plt.xlabel("weight of the cars")
plt.ylabel("Count")
plt.title("Histogram for the weight values")
plt.show()
```

```python
labels = df1['cyl'].value_counts().index
size = df1['cyl'].value_counts().values

plt.pie(size, labels = labels, autopct = '%1.1f%%')
plt.title("cyl pie chart")
plt.show()
```

```python
labels = df1['model'].value_counts().index
size = df1['model'].value_counts().values

plt.pie(size, labels = labels, autopct = '%1.1f%%')
plt.title('model pie chart')
plt.show()
```

### Assignment 5

```python
# Group data

counts = df1.groupby(['am', 'carb']).size().unstack(fill_value=0)
# Area chart
plt.figure(figsize=(8, 5))
plt.stackplot(counts.index, counts.values.T, labels=counts.columns, alpha=0.7)
plt.xlabel('Transmission')
plt.ylabel('Number of Carburetors')
plt.title('Transmission vs Number of Carburetors')
plt.legend(title='Carburetor Count')
plt.show()
```
